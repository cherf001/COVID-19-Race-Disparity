```python
import numpy as np
import pandas as pd
import seaborn as sn
from datetime import datetime as dt
from matplotlib import pyplot as plt
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.width', 200)
```

```python
# Reading File
df = pd.read_csv('COVID_19_Deaths.csv')
```

```python
# Looking at data
df.head()

```

```python
df.info()
```

```python
df.shape
```

```python
# See which columns have missing data 
pd.set_option("max_rows", None)
missing_data = np.sum(df.isnull())
missing_data
```

```python
# dropping the first 9 rows to make data even
df = df.drop([0,1,2,3,4,5,6,7,8])
#Resetting index
df.reset_index(drop=True)
```

```python
## lets see the total cases compared to the ethniticities

plt.figure(figsize=(15,9))
plt.plot(df.DATE,df.African_American, label="African American")
plt.plot(df.DATE,df.White, label="White")
plt.plot(df.DATE,df.Hispanic, label="Hispanic")
plt.plot(df.DATE,df.Asian, label="Asian")
plt.plot(df.DATE,df.Not_Available, label="Not_Available")


plt.xlabel('Date of Diagnosis')
plt.ylabel('Positive Cases By Ethniticy')
plt.title('Comparaitve Analysis By Ethniticity')
plt.legend()
plt.show()
```

On a logarrimthic scale there seem to be an increase in Covid by Whites and followed by African Americans according to this dataset.

```python
#creating a new dataframe to get rid of the OjectID
df2 = df[['DATE', 'African_American', 'White', 'Hispanic', 'Asian', 'Other', 'Not_Available']]
print(df2)
```

```python
# Looking at new dataframe
df2.shape
```

```python
# Creating a Summary Column
df2['Total_Cases'] = df2[['African_American', 'White', 'Hispanic', 'Asian', 'Not_Available']].sum(axis=1)
```

```python
# What are the demographics of counties with the highest percentages of Covid from 4-15-20 to 02-05-21 ?
df2 = pd.DataFrame([np.sum(df2['African_American'])/np.sum(df2['Total_Cases'])*100, 
                       np.sum(df2['White'])/np.sum(df2['Total_Cases'])*100,                   
                       np.sum(df2['Hispanic'])/np.sum(df2['Total_Cases'])*100,
                       np.sum(df2['Asian'])/np.sum(df2['Total_Cases'])*100,
                       np.sum(df2['Other'])/np.sum(df2['Total_Cases'])*100],
                       #np.sum(df2['Not_Available'])/np.sum(df2['Total_Cases'])*100,
                       index = ['African_American','White','Hispanic','Asian','Other']) 
                

df2.plot.bar(color='g', figsize=(10,5), legend=False, rot=0)
plt.title('COVID-19 Deaths by Race')
plt.ylabel('Percent of Covid By Race');
```

Whites are trending above 40% while black are around 39% of Covid according to this data set.
